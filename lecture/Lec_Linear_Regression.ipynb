#Linear Regression


One of the simplest regression problems is fitting a line to data. Regression is when the feature to be predicted contains continuous values.
Regression refers to the process of predicting a dependent variable by analyzing the relationship between other independent variables. Linear regression is a statistical technique used to find the relationship between variables. In an ML context, linear regression finds the relationship between features and a label.


Linear Regression is: A method to help us understand the relationship
between two variables in which the variable to be predicted depends on only one other variable. This is calculated by using the formula that is generally used in calculating the slope of a line.


This is the most basic form of linear regression,

$$y = w_0 + w_1 * x_1$$

In the previous equation,

* $y$ refers to the target variable or the dependenr variable.,

* $x_1$ refers to the independent variable or a feature. It is the input.

* $w_1$ refers to the coefficient that expresses the relationship between $y$ and $x_1$ and is also known as the slope. Also known as the weight of the feature. Weight is the same concept as the slope
 in the algebraic equation for a line. Weight is a parameter of the model and is calculated during training.

* $w_0$ is the constant coefficient, or the intercept. It refers to the constant offset that $y$ will always be with respect to the independent variables. Bias is the same concept as the intercept in the algebraic equation for a line. In ML, bias is sometimes referred to as Bias is a parameter of the model and is calculated during training.



We would like to come up with a linear relationship between the variables,
This is sometimes called a linear mapping but in the one dimensional case it just the equation of a line.

When we fit or train the model we will come up with these parameters. We call this a model,
as it models the relationship between $x$ and $y$.

During training, the model calculates the weight and bias that produce the best model.

For the graph shown below: The linear regression model is:
$$\hat{y} = 1-3x$$
$$ = 1-3(-1)$$

The line will map the dependent variable $x$, to the estimated value of $y$,
the hat on the $y$ indicates it's an estimate. This is known as the $\textit{prediction}$ step.

In linear regression or machine learning in general we have two steps:

* Training Stage: We have a set of training points,
we use these training points to fit or train the model and get parameters. The process of learning parameters for linear regression this is called training. In this example, during training the model calculates the weight and bias that produce the best model.


* Prediction Stage: We then use these parameters in the model to find or predict $\hat{y}$ for new values of $x$.

We now have a model which is a line showing the linear relationship between the input $x$ and continuous valued predictor or output $y$. When $x$ is only one dimension that is there is only one feature, the linear regression is referred to as simple linear regression.

We can use this model to predict any values of y given any x.
Before we discuss how to learn the parameters,
let's see how to make a prediction using some arbitrary values.

## Example-1

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Pounds in 1000s': [3.5, 3.69, 3.44, 3.43, 4.34, 4.42, 2.37],
    'Miles per gallon': [18, 15, 18, 16, 15, 14, 24]
}

# Create DataFrame
df = pd.DataFrame(data)


plt.figure(figsize=(8, 6))
plt.scatter(df['Pounds in 1000s'], df['Miles per gallon'], color='blue', label='Data points')
#plt.plot(df['Pounds in 1000s'], y_pred, color='red', label='Regression line')
plt.title('Pounds in 1000s vs Miles per Gallon with Linear Regression')
plt.xlabel('Pounds in 1000s')
plt.ylabel('Miles per Gallon')
plt.xlim(0, df['Pounds in 1000s'].max() + 1)  # Start x-axis from origin
plt.ylim(0, 35)  # Set y-axis limit to 35
plt.legend()
plt.grid(True)
plt.show()
```

The graph shows Car heaviness (in pounds) versus miles per gallon rating. As a car gets heavier, its miles per gallon rating generally decreases. We could create our own model by drawing a best fit line through the points:

 Picture3.png

In our example, we'd calculate the weight and bias from the line we drew. The bias is 30 (where the line intersects the y-axis), and the weight or slope is calculated as: To calculate the slope of the line passing through the points \( (0, 30) \) and \( (4, 15) \), we can use the slope formula:

$$
w_1 = \frac{y_2 - y_1}{x_2 - x_1}
$$

Where:
- $ (x_1, y_1) = (0, 30) $,
- $ (x_2, y_2) = (4, 15) $.

Substituting these values into the formula:

$
w_1 = \frac{15 - 30}{4 - 0} = \frac{-15}{4} = -3.75
$

So, the slope of the line is $ -3.75 $.

This indicates that for each unit increase in \( x \) (i.e., "Pounds in 1000s"), the value of \( y \) (i.e., "Miles per gallon") decreases by 3.75 units.

* Interpretation: The slope of
$ -3.75 $
  indicates a negative relationship between the two variables: as the pounds increase, the miles per gallon decrease.

The bias  30 means that when the weight of the vehicle ("Pounds in 1000s") is 0, the miles per gallon is 30.

```python
from sklearn.linear_model import LinearRegression

# Reshape data for the model
X = df['Pounds in 1000s'].values.reshape(-1, 1)
y = df['Miles per gallon']

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Plotting the data and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(df['Pounds in 1000s'], df['Miles per gallon'], color='blue', label='Data points')
plt.plot(df['Pounds in 1000s'], y_pred, color='red', label='Regression line')
plt.title('Pounds in 1000s vs Miles per Gallon with Linear Regression')
plt.xlabel('Pounds in 1000s')
plt.ylabel('Miles per Gallon')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Get the slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Print the slope and intercept
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Predictions
y_pred = model.predict(X)
```

```python
# New data values for prediction (e.g., new pounds in 1000s)
new_data = [[3.5], [5.0], [2.0]]  # New data points (reshaped to 2D)

# Predict using the trained model
predictions = model.predict(new_data)

# Print the predicted values for the new data
for i, value in enumerate(new_data):
    print(f"Prediction for {value[0]} Pounds in 1000s: {predictions[i]} Miles per Gallon")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

We can use Scikit-Learn's LinearRegression estimator to fit this data and construct the best-fit line, as shown in the following figure:

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

The slope and intercept of the data are contained in the model's fit parameters, which in Scikit-Learn are always marked by a trailing underscore. Here the relevant parameters are coef_ and intercept_:

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

We see that the results are very close to the values used to generate the data, as we might hope.

The LinearRegression estimator is much more capable than this, howeverin addition to simple straight-line fits, it can also handle multidimensional linear models of the form:
$$y = w_0 + w_1x_1 + w_2 x_2 + \ldots$$

where there are multiple $x$
 values. Geometrically, this is akin to fitting a plane to points in three dimensions, or fitting a hyperplane to points in higher dimensions.

The multidimensional nature of such regressions makes them more difficult to visualize, but we can see one of these fits in action by building some example data, using NumPy's matrix multiplication operator:

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(100, 3)
y = 0.5 + np.dot(X, [1.5, -2., 1.])

model.fit(X, y)
print(model.intercept_)
print(model.coef_)
```

Here the $y$
 data is constructed from a linear combination of three random
 values, and the linear regression recovers the coefficients used to construct the data.

In this way, we can use the single LinearRegression estimator to fit lines, planes, or hyperplanes to our data. It still appears that this approach would be limited to strictly linear relationships between variables, but it turns out we can relax this as well.

## Polynomial Regression##

One trick you can use to adapt linear regression to nonlinear relationships between variables is to transform the data according. The idea is to take the multidimensional linear model

$$y = w_0 + w_1x_1 + w_2 x_2 + \ldots$$

and build the $x_1$, $x_2$, and so on from the single dimensional input $x$. That is, we let $x_p = f_p(x)$ where $f_p(\cdot)$ is some function that transforms our data. For example, if $f_p(x) = x^p$, our model becomes a polynomial regression:

$$y = w_0 + w_1x + w_2 x^2 + w_3 x^3\ldots$$

Notice that this is still a \textbf{linear model}the linearity refers to the fact that the coefficients
 never multiply or divide each other. What we have effectively done is taken our one-dimensional $x$
 values and projected them into a higher dimension, so that a linear fit can fit more complicated relationships between $x$
 and $y$.

```python
from sklearn.preprocessing import PolynomialFeatures
x = np.array([2, 3, 4])
poly = PolynomialFeatures(3, include_bias=False)
poly.fit_transform(x[:, None])
```

The polynomial tranformation has converted our one-dimensional array into a three-dimensional array, where each column contains the exponentiated value. This new, higher-dimensional data representation can then be plugged into a linear regression.

Our linear model, through the use of seventh-order polynomial function, can provide an excellent fit to this nonlinear data!

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns # Import Seaborn
import numpy as np

sns.set_theme(style="whitegrid")
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)
# Plot the data
plt.scatter(x, y) # Scatter plot of the original data
```

```python
from sklearn.preprocessing import PolynomialFeatures # Import PolynomialFeatures
from sklearn.linear_model import LinearRegression # Import LinearRegression
```

```python
# Create polynomial features
poly = PolynomialFeatures(degree=7, include_bias=False) # Create a PolynomialFeatures object with the desired degree
x_poly = poly.fit_transform(x[:, np.newaxis]) # Transform the input features into polynomial features

# Create and fit the polynomial regression model
poly_model = LinearRegression() # Create a LinearRegression object which will be our polynomial model
poly_model.fit(x_poly, y) # Fit the model to the polynomial features and the target variable

# Generate points for plotting the fitted curve
xfit = np.linspace(0, 10, 1000) # Create a range of x values for plotting
xfit_poly = poly.fit_transform(xfit[:, np.newaxis]) # Transform the plotting x values into polynomial features
yfit = poly_model.predict(xfit_poly) # Predict the y values using the fitted model

# Plot the original data and the fitted curve
plt.scatter(x, y) # Scatter plot of the original data
plt.plot(xfit, yfit) # Line plot of the fitted curve
plt.xlabel("x")
plt.ylabel("sin(x) + Noise")
plt.title("Sine Wave with Noise and Polynomial Fit")
plt.show() # Display the plot
```
